```python
import pandas as pd
import numpy as np
import os
```

```python
__path__ = os.getcwd()
```

```python
data_url = os.path.join(
    "_data",
    "21.2.16 MCF10A Stiffness vs HPLM vs HG",
    "2021.2.16_MCF10A_StiffnessvsHPLMvsHG-amounts.csv"
)
```

```python
data = pd.read_csv(
    data_url,
    sep=",")
```

```python
data[[
    "Name",
    "Condition",
    "Exp1", "Exp2", "Exp3"
]]
```

```python
nt_dict = {
    "A": "Adenosine",
    "C": "Cytidine",
    "G": "Guanosine",
    "U": "Uridine"
}

data["Name"] = data["Name"].map(nt_dict).fillna(data["Name"])
```

```python
conditions = data["Condition"].unique().tolist()
replicates = ["Exp1", "Exp2", "Exp3"]
column_names = []

for c in conditions:
    for r in replicates:
        column_names.append(c + "_" + r)
```

```python
data_format = pd.DataFrame(
    index=data["Name"].unique().tolist(),
    columns=column_names)
```

```python
for index, row in data.iterrows():
    for x in replicates:
        data_format.at[row["Name"], row["Condition"] + "_" + x] = row[x]
```

```python
#_data/metabolomics.tsv
data_output_url = os.path.join(
    "_data",
    "21.2.16 MCF10A Stiffness vs HPLM vs HG",
    "2021.2.16_MCF10A_StiffnessvsHPLMvsHG-explode.tsv"
)
```

```python
data_format.to_csv(
    data_output_url,
    sep="\t"
)
```
